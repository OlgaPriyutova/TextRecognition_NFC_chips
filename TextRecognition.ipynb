```python
!sudo apt-get install tesseract-ocr
!pip install pytesseract

!pip show pytesseract

import matplotlib.pyplot as plt
from PIL import Image
import io
import pytesseract

def extract_text_from_image_direct(image_path):
    try:
        with Image.open(image_path) as img:
            # Display the image
            plt.imshow(img)
            plt.axis('off')  # Turn off axis numbers
            plt.show()

            # Process the image for OCR
            img_byte_arr = io.BytesIO()
            img.save(img_byte_arr, format='JPEG')  # Use 'PNG' if the image is in PNG format
            img_byte_arr = img_byte_arr.getvalue()
            text = pytesseract.image_to_string(Image.open(io.BytesIO(img_byte_arr)))
            return text
    except IOError:
        return "Error: Could not process the image."

extracted_text = extract_text_from_image_direct('/content/pass.jpg')
print("Extracted Text:\n", extracted_text)
```
